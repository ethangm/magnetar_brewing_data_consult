### one hot encoding

```python
def style_encoding(df: pd.DataFrame, col: str) -> pd.DataFrame:
    hot_encoded = pd.get_dummies(df[col], dtype=np.int64, prefix=col)
    df_new = pd.concat([df.drop(col, axis=1), hot_encoded], axis=1)
    return df_new
```

```python
df_encoded = style_encoding(df, "beer_style")
#df_encoded = style_encoding(df_encoded, "brewery_name")
df_encoded.head(10)
```

^ Why are these different?

dataframe becomes too large if I one hot encode breweries. I will try target encoding (w.r.t. overall_rating) instead

```python
from sklearn.preprocessing import TargetEncoder
def target_encoding(df: pd.DataFrame, col: str, target_col: str, smooth: bool = True) -> pd.DataFrame:
    if smooth:
        targ_enc = TargetEncoder(shuffle=False)
    else:
        targ_enc = TargetEncoder(shuffle=False, smooth=0)
    
    X_trans = targ_enc.fit_transform(df.loc[:, [col]], df[target_col])

    df_new = pd.concat([df.drop(col, axis=1), pd.DataFrame(X_trans, df.index, [f"{col}_encoded"])], axis=1)
    return df_new
```

```python
df_targ_encoded = target_encoding(df_encoded, "brewery_name", "review_overall")
df_targ_encoded.head(10)
```

```python
def beerid_map(df: pd.DataFrame) -> dict:
    temp = df[["beer_name", "brewery_name", "beer_beerid"]].drop_duplicates()
    beer_dict = {}
    for _, row in temp.iterrows():
        beer_dict[row["beer_beerid"]] = (row["beer_name"], row["brewery_name"])
    
    return beer_dict
```

```python
temp = beerid_map(df)
```
