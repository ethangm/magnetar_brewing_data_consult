```python
def style_encoding(df: pd.DataFrame, col: str) -> pd.DataFrame:
    hot_encoded = pd.get_dummies(df[col], dtype=np.int64, prefix=col)
    df_new = pd.concat([df.drop(col, axis=1), hot_encoded], axis=1)
    return df_new
```

```python
from sklearn.preprocessing import TargetEncoder
def target_encoding(df: pd.DataFrame, col: str, target_col: str, smooth: bool = True) -> pd.DataFrame:
    if smooth:
        targ_enc = TargetEncoder(shuffle=False)
    else:
        targ_enc = TargetEncoder(shuffle=False, smooth=0)

    X_trans = targ_enc.fit_transform(df.loc[:, [col]], df[target_col])

    df_new = pd.concat([df.drop(col, axis=1), pd.DataFrame(X_trans, df.index, [f"{col}_encoded"])], axis=1)
    return df_new
```

```python
from numpy.linalg import norm   # keep it simple to one review per user

def norm_score(df: pd.DataFrame):
    # inplace
    review_cols = ["review_overall", "review_taste", "review_palate", "review_aroma", "review_appearance"]
    for ind, row in df.iterrows():
        df.loc[ind, "norm_reviews"] = norm(row[review_cols].values)




# want output of form x rows of users, y columns of beers, z rating criteria
def user_item_matrixer(df: pd.DataFrame):
    # reviews are normalized by user rating average

    num_beers = df.beer_beerid.nunique()
    num_users = df.review_profilename.nunique()
    beers = df.beer_beerid.unique()
    users = df.review_profilename.unique()

    ui_matrix = pd.DataFrame(index=users, columns=beers)
    norm_score(df)

    avg_user_ratings = df.groupby("review_profilename")["norm_reviews"].mean()

    for ind, row in df.iterrows():
        score = norm(row[review_cols].values) - avg_user_ratings[row["review_profilename"]]

        ui_matrix.loc[row["review_profilename"], row["beer_beerid"]] = score

    avg_beer_rating = df.groupby("beer_beerid")["norm_reviews"].mean()

    return ui_matrix, avg_user_ratings, avg_beer_rating
```

```python
from numpy import dot

# cosine similarity = a dot b / (norm(a) * norm(b))
def cossim(a, b):   # nans were causing nan output
    a_num = np.array(a, dtype=np.float_)
    b_num = np.array(b, dtype=np.float_)

    a_num = np.nan_to_num(a_num, copy=False)
    b_num = np.nan_to_num(b_num, copy=False)

    temp = dot(a_num, b_num)
    if temp == 0:
        return 0
    else:
        return temp / (norm(a_num) * norm(b_num))
```

```python
def sim_matrix(df: pd.DataFrame) -> pd.DataFrame:
    item_matrix = pd.DataFrame(index=df.columns, columns=df.columns)
    for i in df.columns:
        for j in df.columns:
            if i == j:
                item_matrix.loc[i, j] = 1
            else:
                temp = cossim(df.loc[:, i], df.loc[:, j])
                item_matrix.loc[i, j] = temp
    return item_matrix
```

```python
class Recommender:

    def __init__(self, item_matrix: pd.DataFrame, beerid_dict: dict, avg_beer_ratings):
        self.item_matrix = item_matrix
        self.beerid_dict = beerid_dict
        self.beer_avgs = avg_beer_ratings
        self.recommendations = pd.DataFrame(columns=item_matrix.columns)

    def recommend(self, reviews: dict, user=str):
        # for every item not reviewed, multiply rating of an item review by
        # the similarity to item not reviewed, divided by similarity
        # reviews = {beer id: list of ratings}

        avg = self.user_avg(reviews)

        calculated_ratings = []
        for i in self.item_matrix.columns:
            if i not in reviews.keys():     # don't recommend a beer they've already had
                calculated_ratings.append(self._rec_helper(reviews, i, avg))
            else:
                calculated_ratings.append(np.NaN)

        self.recommendations.loc[user] = calculated_ratings

        # sort recommendations by value, take top three, input corresponding column into beerid->name dict
        temp = self.recommendations.loc[user].sort_values(ascending=False)[:3].index
        print(f"Top 3 Beers to Try!\n1. {self.beerid_dict[temp[0]]}\n2. {self.beerid_dict[temp[1]]}\n3. {self.beerid_dict[temp[2]]}")
        #print(self.recommendations.loc[user].sort_values(ascending=False)[:30])

    def _rec_helper(self, reviews, beer, avg) -> np.float_:
        # TODO: calculate the rating/similarity score for beer

        numerator = 0
        denominator = 0
        for i, ratings in reviews.items():
            score = norm(ratings)
            if i not in self.item_matrix.columns:
                # TODO
                # live updating of item_matrix
                continue
            else:
                numerator += self.item_matrix.loc[beer, i] * (score - self.beer_avgs[i])
                denominator += self.item_matrix.loc[beer, i]

        return (numerator / denominator) + avg

    @staticmethod
    def user_avg(reviews):
        temp = []
        for i in reviews.values():
            temp.append(norm(i))
        return np.mean(temp)
```
